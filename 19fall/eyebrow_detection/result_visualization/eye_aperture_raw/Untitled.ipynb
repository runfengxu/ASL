```python
files = os.listdir('eye_aperture_raw')
for file in files[:1]:
    data = pd.read_csv('eye_aperture_raw/'+file, header = None)
    data[0] = data[0].map(lambda x: x.split('frameID: ')[-1])
    data[1] = data[1].map(lambda x: x.split('eyeAperture: ')[-1])
    data[2] = data[2].map(lambda x: x.split('pitch: ')[-1])
    data[3] = data[3].map(lambda x: x.split('yaw: ')[-1])
    data[4] = data[4].map(lambda x: x.split('roll: ')[-1])
```
