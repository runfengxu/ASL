```python
from xml.dom.minidom import parse
import xml.dom.minidom
import json
import numpy as np
import scipy.io as scio
import os
import math
from os import path



DOMTree = xml.dom.minidom.parse('781_12582.xml')
collection= DOMTree.documentElement
```

```python
collections = collection.getElementsByTagName('COLLECTION')
```

```python
s={'conditional/when','negative','relative','rhetorical question','topic/focus','wh question','yes-no question'}
l={}
for collection in collections:
    mediafile = collection.getElementsByTagName('MEDIA-FILES')[0].getElementsByTagName('MEDIA-FILE')[0].getAttribute('FILE-NAME')
    l[mediafile]=[]
    utterances = collection.getElementsByTagName('UTTERANCE')
    for utterance in utterances:
        ut={}
        ut['sign']=[]
        ut['special']=[]
        start =int(utterance.getAttribute('START_FRAME'))/1001
        end = int(utterance.getAttribute('END_FRAME'))/1001
        ut['start']=int(start)
        ut['end'] = int(end)
        signs = utterance.getElementsByTagName('SIGN')
        for sign in signs:
            label = sign.getElementsByTagName('LABEL')[0].childNodes[0].nodeValue
            start = sign.getElementsByTagName('DOMINANT_HAND')[0].getAttribute('START_FRAME')
            end = sign.getElementsByTagName('DOMINANT_HAND')[0].getAttribute('END_FRAME')
            start = int(int(start)/1001)
            end = int(int(end)/1001)
            ut['sign'].append((label,start,end))
        specials = utterance.getElementsByTagName('NON_MANUAL')
        for special in specials:
            label = special.getElementsByTagName('LABEL')[0].childNodes[0].nodeValue
            
            
            if label[1:-1] in s:
                print(label)
                start = int(int(special.getAttribute('START_FRAME'))/1001)
                end = int(int(special.getAttribute('END_FRAME'))/1001)
                ut['special'].append((label,start,end))
        l[mediafile].append(ut)
        
        
        
```

```python
for key in l.keys():
    if not os.path.exists('sign/'+key):
        os.mkdir('sign/'+key)
    for ut in l[key]:
        scio.savemat('sign/'+key+'/'+str(ut['start'])+'to'+str(ut['end']),{'sign':ut['sign'],'special':ut['special']})
```
