```python
files = os.listdir('test')
utterance = {}
for file in files:
    if file.split('cam2')[0] not in utterance.keys():
        utterance[file.split('cam2')[0]]=[]
    utterance[file.split('cam2')[0]].append((file.split('_')[-1].split('to')[0],file.split('to')[-1].split('.mat')[0]))
    
```

```python
data = scio.loadmat('test2.mat')
test=[]
for i in data['img_file']:
    video = i[0][7:].split('cam2')[0]
    frame = i[0][7:].split('frame')[1].split('.jpg')[0]
    for j in utterance[video]:
        if int(j[0])<=int(frame) and int(j[1])>=int(frame):
            test.append((video,j[0],j[1]))
            
```
