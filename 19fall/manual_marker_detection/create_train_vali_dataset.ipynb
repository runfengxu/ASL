```python
import pandas as pd
import scipy.io as scio
import os
```

```python
files = os.listdir('dataset')
time0=0
time1 = 0
for file in files:
    if file[-4:]=='.mat'  and '_'.join(file.split('_')[:-2])=='2013-06-27_CB_0112-cam1-for-ss3':
        time0 +=eval( file.split('_')[-1].split('to')[-1][:-4])-eval(file.split('_')[-1].split('to')[0])
        data = scio.loadmat('dataset/'+file)['sign'][0]
        for i in range(len(data)):
            if data[i]==1:
                time1+=1

print(time1/time0)
```

```python
sign1=[]
img1=[]
files= os.listdir('dataset')
random.shuffle(files)
for file in files[:650]:
    print(file)
    if file[-4:]=='.mat':
        data = scio.loadmat('dataset/'+file)
        #print(data)
        sign1 +=data['sign'][0].tolist()
        img1 +=data['img_file'].tolist()
        shutil.copyfile('dataset/'+file, 'dataset/train/'+file)
    
    
sign2=[]
img2=[]
for file in files[651:]:
    print(file)
    if file[-4:]=='.mat':
        data = scio.loadmat('dataset/'+file)
        #print(data)
        sign2 +=data['sign'][0].tolist()
        img2 +=data['img_file'].tolist()
        shutil.copyfile('dataset/'+file, 'dataset/vali/'+file)
```

```python
scio.savemat('dataset/train.mat',{'sign':sign1,'img_file':img1})
scio.savemat('dataset/validation.mat',{'sign':sign2,'img_file':img2})
```
